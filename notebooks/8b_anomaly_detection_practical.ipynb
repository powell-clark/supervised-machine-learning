```python
import numpy as npimport matplotlib.pyplot as pltfrom sklearn.ensemble import IsolationForestfrom sklearn.svm import OneClassSVMfrom sklearn.covariance import EllipticEnvelopefrom sklearn.datasets import make_classificationfrom sklearn.metrics import classification_report, roc_auc_scorenp.random.seed(42)
```

## Generate Anomaly Data

```python
# Normal datanormal = np.random.normal(0, 1, (1000, 2))# Anomaliesanomalies = np.random.uniform(-4, 4, (50, 2))X = np.vstack([normal, anomalies])y_true = np.array([0]*1000 + [1]*50)  # 0=normal, 1=anomalyplt.figure(figsize=(10, 6))plt.scatter(X[y_true==0, 0], X[y_true==0, 1], alpha=0.5, label='Normal', s=20)plt.scatter(X[y_true==1, 0], X[y_true==1, 1], c='red', label='Anomaly', s=50, marker='x')plt.legend()plt.title('Anomaly Detection Problem')plt.show()
```

## Compare Methods

```python
methods = {    'Isolation Forest': IsolationForest(contamination=0.05, random_state=42),    'One-Class SVM': OneClassSVM(gamma='auto', nu=0.05),    'Elliptic Envelope': EllipticEnvelope(contamination=0.05, random_state=42)}for name, model in methods.items():    y_pred = model.fit_predict(X)    y_pred = np.where(y_pred == -1, 1, 0)    print(f'{name}:')    print(classification_report(y_true, y_pred, target_names=['Normal', 'Anomaly']))    print()
```

## Production Recommendations**Isolation Forest:**- ✅ Default choice for most cases- ✅ Fast, scalable- ✅ Works with high dimensions**Use Case: Credit Card Fraud**- Train on normal transactions- Flag unusual patterns- Set contamination based on expected fraud rate
